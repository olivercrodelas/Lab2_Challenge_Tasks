```python
# --- APPENDIX TASK 1: REDACTION PROTOCOL ---
# Objective: Obscure identity by replacing vowels with Seed Number.

SURNAME = "RODELAS"
SEED_NUM = 2
redacted_name = ""

print(f"Original Identity: {SURNAME}")

# Loop through each character
for char in SURNAME:
    if char in "AEIOU":       # Check if character is a vowel
        redacted_name += str(SEED_NUM)
    else:                     # If consonant, keep it
        redacted_name += char

print(f"Redacted Identity: {redacted_name}")
```

```python
# --- APPENDIX TASK 2: SECURE INPUT LOOP ---
# Objective: Lock system until correct credentials are provided.

CORRECT_PASS = "TUPM-25-0512"  # Your Student ID
attempts = 0
access_granted = False

print("--- SECURITY GATE ACTIVE ---")

while not access_granted:
    user_input = input("Enter Admin Password: ")
    attempts += 1
    
    if user_input == CORRECT_PASS:
        print(f"ACCESS GRANTED. Welcome, Engineer.")
        print(f"Total Attempts: {attempts}")
        access_granted = True
    else:
        print("ACCESS DENIED. Incorrect Password.")
```

```python
# --- APPENDIX TASK 3: FORENSIC VERIFICATION ---
# Formula: (ASCII of First Letter * Seed) - Total Loops

ENGINEER = "Rodelas"
FIRST_LETTER = ENGINEER[0]   # Extracts 'R'
ASCII_VAL = ord(FIRST_LETTER) # Converts 'R' to 82
SEED_NUM = 2
LOOPS = 5                    # From Task 12

# Calculate the Hash
verification_hash = (ASCII_VAL * SEED_NUM) - LOOPS

print(f"First Letter: {FIRST_LETTER} (ASCII: {ASCII_VAL})")
print(f"Seed: {SEED_NUM}")
print(f"Loops Deducted: {LOOPS}")
print(f"FINAL FORENSIC HASH: {verification_hash}")
```
